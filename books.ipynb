### Computing cosine-similarity using authors, publication year, title, Language_code, average rating, ratings_count and tag_name

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from scipy.sparse import coo_matrix, hstack
```

```python
f = (pd.Series(books[['authors','title']]
                .fillna('')
                .values.tolist()
                ).str.join(' '))
```

```python
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(f.astype('U'))
```

```python
#tfidf_matrix = hstack([tfidf_matrix,np.array(final_df['average_rating'].values[:2000]).reshape(-1,1)])
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim
```

```python
titles = books['original_title']
indices = pd.Series(books.index, index=books['title'])

# Function that get book recommendations based on the cosine similarity score of book authors
def authors_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:21]
    book_indices = [i[0] for i in sim_scores]
    rec = []
    for j in book_indices:
        dic = {}
        dic["title"] = books['title'].iloc[j]
        dic["img"] = books['image_url'].iloc[j]
        dic["year"] = books['original_publication_year'].iloc[j]
        dic["rating"] = books['average_rating'].iloc[j]
        dic["author"] = books['authors'].iloc[j].replace('  ','') 
        rec.append(dic)
#     print(rec)
#     img_urls = [books['image_url'].iloc[i] for i in book_indices]
#     year = [books['original_publication_year'].iloc[i] for i in book_indices]
#     rating = [books['average_rating'].iloc[i] for i in book_indices]
#     authors = [books['authors'].iloc[i].replace('  ','') for i in book_indices]
    return rec
```
